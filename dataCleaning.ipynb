This notebook is used to show how we can clean and process data for further analysis. 

```python
#  importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data= pd.read_csv('Walmart_sales.csv')
```

```python
data.head()
```

```python
data_filtered = data.drop(['CPI'], axis=1)
```

```python
data_filtered.head()
```

```python
filtered_data = data[data['Temperature'] > 40]
filtered_data
```

```python
filtered_data = filtered_data.query("Fuel_Price > 3 and Store != 1")
filtered_data
```

```python
#  removing duplicates
```

```python
duplicates = data.duplicated()
duplicates
```

```python
data = data.drop_duplicates()
data.head()
```

```python
data = data[~data.duplicated(subset=['column1', 'column2'])]
data
```

```python
#    Handling missing data
missing_data = data.isnull()
missing_data
```

```python
# Drop rows with any missing value
data.dropna()
data
```

```python
# Drop columns with any missing value
data.dropna(axis='columns')
data
```

```python
data.fillna(0)
```

```python
#  fill using summary stastics
mean_value = data['CPI'].mean()
data['CPI'].fillna(mean_value, inplace=True)
```

```python
# Forward fill missing values
data.fillna(method='ffill')

# Backward fill missing values
data.fillna(method='bfill')
```

```python
#   outlier detection
```

```python
plt.hist(data['Temperature'], bins=10)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Values')
plt.show()
```

```python
plt.scatter(data.index, data['Temperature'])
plt.xlabel('Index')
plt.ylabel('Value')
plt.title('Scatter Plot of Values')
plt.show()
```

```python
# Plot box plot
import seaborn as sns
sns.boxplot(data=data, x='weekly_sales')
plt.xlabel('Weekly_Sales')
plt.title('Box Plot of Values')
plt.show()
```

```python
#  inclistences in column names fix by this
```

```python
data.columns = [column.lower().replace(' ', '_') for column in data.columns]
```

```python
data.head()
```

```python
# Standarzing and normalizing 
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
# Scale the data

data= data.drop(['date'], axis=1)

modified_data = data.copy()

scaler = MinMaxScaler()  
scaled_values = scaler.fit_transform(data)
modified_data[data.columns] = scaled_values

# Standardize the scaled data
standardizer = StandardScaler()
standardized_values = standardizer.fit_transform(scaled_values)
modified_data[data.columns] = standardized_values

# Save the scaled and standardized DataFrame to a new CSV file
modified_data.to_csv('dataNew.csv', index=False)
```
